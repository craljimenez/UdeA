### 3.4 Scope (Alcance de variables) en Python

En Python, el **scope** o alcance determina desde dónde se puede acceder a una variable o nombre definido en el programa.

#### Tipos de scopes principales:
- **Local:** Dentro de una función.
- **Global:** Fuera de todas las funciones, en el nivel principal del script o notebook.
- **Built-in:** Nombres reservados de Python (como `print()`, `len()`).

Recordatorio clave:
- Las variables definidas dentro de una función solo existen dentro de ella (local).
- Las variables definidas fuera de todas las funciones son accesibles globalmente (global).

as.png

#### Ejemplo 1: Scope global

Las variables definidas fuera de cualquier función son globales.



````python
x = 10  # Variable global
def mostrar_x():
    print("Dentro de la función, x =", x)

mostrar_x()
print("Fuera de la función, x =", x)
````

Las variables definidas dentro de una función son locales y no existen fuera de ella.

#### Ejemplo 3: Modificar una variable global dentro de una función

Para modificar una variable global dentro de una función, usamos la palabra clave `global`.


````python
z = 20

print("Antes de función, z=",z)

def cambiar_z():
    global z
    z = 99
    print("Dentro de la función, z =", z)

cambiar_z()
print("Fuera de la función, z =", z)
````

#### Ejemplo 5: Anidamiento de funciones y el scope `nonlocal`

Cuando tenemos funciones anidadas, podemos usar `nonlocal` para modificar una variable en el scope de la función externa.


````python
contador = 5 # variable Global
def externa():
    contador = 0 # Variable Local
    print("Dentro de externa, antes de interna,contador =", contador)

    def interna():
        nonlocal contador # VAriable Local de la función externa
        contador = 50
        contador += 1 # contador = contador + 1
        print("Dentro de interna, contador =", contador)

    interna()
    print("Dentro de externa, contador =", contador)

print("Antes de función, contador",contador)
externa()
print("Después de función, contador",contador)
````

Escribe una función llamada `contar_vocales(cadena)` que reciba una cadena de texto y devuelva el número total de vocales (`a`, `e`, `i`, `o`, `u`) en la cadena. Usa un bucle `for` para recorrer cada carácter.


````python
def contar_vocales(cadena_texto):
    vocales = "aeiouAEIOUáéíóúÁÉÍÓÚ"
    contador_vocales = 0
    for char in cadena_texto:
        for vocal in vocales:
            if char == vocal:
                contador_vocales += 1
    return contador_vocales

contar_vocales("Cristian")
````

````python
from time import time
def contar_vocales(cadena_texto):
    vocales = "aeiouAEIOUáéíóúÁÉÍÓÚ"
    contador_vocales = 0
    num_vocales = len(vocales)
    N = len(cadena_texto)
    # i,j = 0,0
    i = 0

    while i < N:
        j = 0
        while j < num_vocales:
            if cadena_texto[i] == vocales[j]:
                contador_vocales += 1
                # break
            j += 1

        i += 1

    return contador_vocales
tic = time()
contar_vocales("Cristian")
toc = time()

print("Tiempo ejecución: ",toc - tic)
````

Escribir una función que verifique si una palabra termina en vocal, y luego usarla para contar cuántas palabras de una cadena cumplen esta condición.

````python
ejemplo = "Hola Mundo cómo esta usted manzana"
def contar_palabras_finalizadas_en_vocal(texto):
    """
    Documentación de la función
    Entradas:
        texto (str): texto para ver la cantidad de palabras
                     que finalizan con vocal
    Salidas:
        contador_palabras (int): El numero de palabras de la entrada texto
                                 que finaliza con vocal
    """
    palabras = texto.split(" ") # Lista
    N = len(palabras)
    i = 0
    contador_palabras = 0
    while i<N:
        palabra = palabras[i] # palabra es str, es una cadena de caracteres
        num_vocales = contar_vocales(palabra[-1])
        if num_vocales > 0:
            contador_palabras += 1
        i += 1

    return contador_palabras

contar_palabras_finalizadas_en_vocal(ejemplo)
````

# Aliasing
En Python, cuando asignas una variable a otra variable que ya referencia un objeto, no estás creando una copia de ese objeto. En su lugar, ambas variables terminan referenciando el mismo objeto en memoria. Este fenómeno se conoce como **aliasing**.

Imagina que tienes un nombre para una persona y luego le asignas otro nombre. Ambos nombres se refieren a la misma persona. En Python, las variables son como esos nombres y los objetos en memoria son como las personas.

````python
lista1 = [1, 2, 3]
lista2 = lista1 # Aquí ocurre el aliasing
````

````python
id(lista1) == id(lista2)
````

En este caso, `lista1` y `lista2` no son dos listas separadas con los mismos elementos. Ambas variables apuntan a la misma lista en la memoria.

**Consecuencias del Aliasing**:

La principal consecuencia del aliasing es que si modificas el objeto a través de una de las variables, los cambios se reflejarán cuando accedas al objeto a través de la otra variable, porque ambas están referenciando el mismo objeto.

````python
lista1.append(4)

print(lista1)
print(lista2)
````

````python
lista2.remove(2)
print("lista1:",lista1)
print("lista2:",lista2)
````

````python
lista3 = lista1.copy() # Evitar aliasing, creando una copia.
````

````python
lista3.insert(1,2)
print("lista1:",lista1)
print("lista3:",lista3)
````

## ¿Qué pasa cuando usamos aliasing en el cuerpo de una función?

````python
def agregar_elemento(lista,elem):
    """
    Agrega un elem al final de la lista llamada lista.
    """
    lista_aliasing = lista # Aliasing
    lista_aliasing.append(elem)
    print("lista_aliasing:",lista_aliasing)
    return lista_aliasing

print("lista1 antes de usar función con asliasing:",lista1)

lista3 = agregar_elemento(lista1,10)

print("lista1 después de usar función con asliasing:",lista1)
print("lista3:",lista3)
````

````python
def agregar_elemento(lista,elem):
    """
    Agregar elem a lista
    """
    lista.append(elem)

print("lista1 antes de usar función:",lista1)

agregar_elemento(lista1,20)

print("lista1 después de usar función:",lista1)
````

````python
def agregar_elemento(lista,elem):
    """
    Agregar elem a la lista y retornar una copia de esta sin alterar la original
    """
    lista_aux = lista.copy() # Evitando aliasing
    lista_aux.append(elem)
    return lista_aux

print("lista1 antes de usar función:",lista1)

lista4 = agregar_elemento(lista1,30)

print("lista1 después de usar función:",lista1)
print("lista4:",lista4)
````

Puedes crear una matriz de diferentes maneras:

- **Usando bucles**: Puedes usar bucles anidados para construir la matriz, especialmente si necesitas inicializarla con valores específicos o si el tamaño de la matriz es variable.

````python
def create_matrix_zeros(filas,columnas):
    """
    crea un amatriz de ceros de tamaño filas x columnas
    """
    matriz_ceros = [] # list()

    for i in range(filas):
        fila = []
        for j in range(columnas):
            fila.append(0) # Inicializamos con ceros
        matriz_ceros.append(fila)
    return matriz_ceros


matriz_ceros = create_matrix_zeros(3,4)
print(matriz_ceros)
````

## Producto Punto

El producto punto es una operación que toma dos vectores de igual longitud y devuelve un único valor escalar (un número). Conceptualmente, mide "cuánto" dos vectores apuntan en la misma dirección.

Definición:

Dados dos vectores $\mathbf{a} = [a_1, a_2, \dots, a_n]$ y $\mathbf{b} = [b_1, b_2, \dots, b_n]$ de longitud $n$, su producto punto se define como:

$\mathbf{a} \cdot \mathbf{b} = a_1b_1 + a_2b_2 + \dots + a_nb_n = \sum_{i=1}^{n} a_ib_i$

En otras palabras, multiplicas los elementos correspondientes de los dos vectores y luego sumas todos esos productos.

Implementación en Python con Listas:

Podemos representar vectores como listas simples en Python. Aquí tienes una función para calcular el producto punto:

````python
def producto_punto(vectorA, vectorB):
    """
    Calcula el producto punto de dos vectores representados como listas.

    Args:
        vectorA: La primera lista (vector).
        vectorB: La segunda lista (vector).

    Returns:
        El valor escalar del producto punto,
        o un mensaje de error si los vectores no tienen la misma longitud.
    """

    # Verificar si los vectores tienen la misma longitud
    if len(vectorA) != len(vectorB):
        return "No se puede calcular el producto punto: los vectores deben tener la misma longitud."

    resultado = 0
    for i in range(len(vectorA)):
        resultado += vectorA[i] * vectorB[i]

    return resultado

# Ejemplo de uso:
vector1 = [1, 0, 1]
vector2 = [0, 1, 0]

resultado_producto_punto = producto_punto(vector1, vector2)
print(f"El producto punto de {vector1} y {vector2} es: {resultado_producto_punto}")

vector3 = [10, -2]
vector4 = [5, 3]

resultado_producto_punto2 = producto_punto(vector3, vector4)
print(f"El producto punto de {vector3} y {vector4} es: {resultado_producto_punto2}")

vector5 = [1, 2]
vector6 = [3, 4, 5]

resultado_producto_punto3 = producto_punto(vector5, vector6)
print(resultado_producto_punto3)
````

## Multiplicación
El resultado de la multiplicación de dos matrices se obtiene mediante el producto punto. En particular, el elemento $[i,j]$ de la matriz resultante se calcula como el producto punto entre la fila $i$-ésima de la primera matriz y la columna $j$-ésima de la segunda matriz.

$$
\begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}
\end{bmatrix}
⋅\begin{bmatrix}
b_{11} & b_{12} \\
b_{21} & b_{22}
\end{bmatrix}
=
\begin{bmatrix}
a_{11}b_{11} + a_{12}b_{21} & a_{11}b_{12} + a_{12}b_{22} \\
a_{21}b_{11} + a_{22}b_{21} & a_{21}b_{12} + a_{22}b_{22}
\end{bmatrix}
$$


````python
M = [[1,2,3],
    [4,5,6],
    [7,8,9]
    ]
j = 0
for i,row in enumerate(M):
    print(f"row_{i}[{j}]:",row[j])
````

````python
M = [[1,2,3],
    [4,5,6],
    [7,8,9]
    ]
def extract_column(M,j=0):
    col_j = []
    for row in M:
        col_j.append(row[j])
    return col_j

print(extract_column(M,1))
````

````python
def multiplicar_matrices(matrizA, matrizB):
    """
    Multiplica dos matrices representadas como listas anidadas.

    Args:
        matrizA: La primera matriz (lista de listas).
        matrizB: La segunda matriz (lista de listas).

    Returns:
        Una nueva matriz resultante de la multiplicación,
        o un mensaje de error si la multiplicación no es posible.
    """

    # Obtener las dimensiones de las matrices
    filasA = len(matrizA)
    columnasA = len(matrizA[0])
    filasB = len(matrizB)
    columnasB = len(matrizB[0])

    # Verificar si la multiplicación es posible
    if columnasA != filasB:
        return "No se pueden multiplicar las matrices: el número de columnas de la primera matriz debe ser igual al número de filas de la segunda."

    # Inicializar la matriz resultante con ceros
    matriz_resultado = create_matrix_zeros(filasA,columnasB)

    # Realizar la multiplicación
    for i in range(filasA): # recorro Filas de la primera matriz
        for j in range(columnasB): #recorro columnas de la segunda matriz
            fila_i_A = matrizA[i]
            columna_j_B = extract_column(matrizB,j)
            # Calcular el elemento en la posición (i, j) de la matriz resultado
            matriz_resultado[i][j] = producto_punto(fila_i_A,columna_j_B)

    return matriz_resultado

# Ejemplo de uso:
matriz1 = [
    [1, 2 , 4],
    [3, 4, 5]
]

matriz2 = [
    [5, 6],
    [7, 8],
    [1,-1]
]

resultado = multiplicar_matrices(matriz1, matriz2)
print(resultado)

matriz3 = [
    [1, 2, 3],
    [4, 5, 6]
]

matriz4 = [
    [7, 8],
    [9, 10],
    [11, 12]
]

resultado2 = multiplicar_matrices(matriz3, matriz4)
print(resultado2)

matriz5 = [
    [1, 2]
]

matriz6 = [
    [3, 4]
]

resultado3 = multiplicar_matrices(matriz5, matriz6)
print(resultado3)
````

# Tuplas en Python 🔐

Una **tupla** es una estructura de datos similar a una lista, pero con una diferencia fundamental: **es inmutable**, lo que significa que **no puede ser modificada** después de su creación.

Su definición de una tupla se puede hacer:

1. Por medio de la función:
> $tuple(elem_1,elem_2,\dots,elem_N)$
2. Por medio de parentesis `()` y sus elementos separados por comma (`,`)
> $tupla = (elem_1,elem_2,\dots,elem_N)$

Si python detecta parantesis y entre ellos una comma, lo interpretara cómo una tupla.

✅ **Similitudes con listas**:
- Se definen como secuencias ordenadas.
- Soportan indexación y slicing.
- Pueden contener elementos de cualquier tipo.



````python
tupla = (10,20,30) # tuple(10,20,30)
print("tupla:",tupla)
print("tipo de datos de una tupla (type(tupla)):",type(tupla))
print("Acceso a un elemento --> tupla[0]:",tupla[0]) #
print("Realización de Slicing/rebanado --> tupla[1:]:",tupla[1:]) #
````

## Mutabilidad y aliasing 🔄

- **Aliasing** es cuando dos variables apuntan al mismo objeto en memoria.

## Busqueda y Ordenamiento

### LinearSearch
Complejidad: $\mathcal{O}(len(L))$

````python
def linSearch(L,e):
    for k in range(len(L)):
        if L[k] == e:
            return True
    return False
````

````python
L = [4,5,1,2,-4,6,3]
e = -11
linSearch(L,e)
````

### OrdSearch
Busqueda desde array ordenada.

- **Complejidad en Peor de los casos**: $\mathcal{O}(len(L))$.
- **Complejidad Caso Promedio**: $\mathcal{O}\left(\frac{len(L)}{2}\right)$.

````python
def ordSearch(L,e):
    for k in range(len(L)):
        if L[k] == e:
            return True
        elif L[k] > e:
            return False
    return False
````

````python
L = [2,3,6,10,40]
e = 5
ordSearch(L,e)
````

### Binary Search

**Asume que lel arreglo esta ordenado de forma ascendente**


**Complejidad:** $\mathcal{O}(\log_2\{len(L)\})$

````python
def binSearch(L,e):
    low = 0
    high = len(L) - 1
    while low <= high:
        mid = (low + high)//2
        if e > L[mid]:
            low = mid + 1
        elif e < L[mid]:
            high = mid - 1
        else:
            return True
    return False
````

### Ordenamiento por selección

Se asume que L es un elemento cuyos elementos se pueden comparar con el operador relacional $<$.

````python
def selSort(L):
    sufStart = 0 # 1
    N = len(L) # 1
    print(f"L\t\t\tlen(L)\ti\tL[i]\tsurfStart | L[sufStart]")
    while sufStart < N: # O(len(L))
        for i in range(sufStart+1,N): # O(len(L))
            print(f"{L}\t{N}\t{i}\t{L[i]}\t{sufStart}\t\t{L[sufStart]}")
            if L[i] < L[sufStart]: # 4
                temp = L[i] # 1
                L[i] = L[sufStart] # 1
                L[sufStart] = temp # 1
        sufStart += 1 # 2
    print(f"{L}\t{N}\t{i}\t{L[i-1]}\t{sufStart}\t\t{L[sufStart-1]}")

L = [5,2,10,1,9]

selSort(L)
````

### Burble Sort

**¿Cómo funciona?**
- Compara pares de elementos adyacentes.

- Si están en el orden incorrecto, los intercambia.

- Cada pasada "burbujea" el mayor valor restante al final de la lista.

- Repite esto hasta que la lista esté ordenada.

**Complejidad temporal:**
Peor y promedio: $\mathcal{O}(len(L)^2)$

````python
def bubble_sort(L):
    n = len(L)
    print(f"i\tj\tL[j]\tL[j+1]\tL")
    for i in range(n):
        for j in range(0, n - i - 1):
            print(f"{i}\t{j}\t{L[j]}\t{L[j+1]}\t{L}")
            if L[j] > L[j + 1]:
                L[j], L[j + 1] = L[j + 1], L[j]

L = [5,2,10,1,9]

bubble_sort(L)
````

### Ordenamiento por Mezcla

Este algoritmo, creado por John von Neumann en 1945, utiliza una idea general que está presente en muchos algoritmos eficientes: divide y vencerás. El algoritmo parte de observar que dos listas ordenadas pueden mezclarse, generando una nueva lista ordenada en tiempo lineal.

**Asume que ambas listas son listas ordenadas. Retorna una lista ordenada con a union de ambas listas.**

````python
def merge(left,right):
    result = []
    i = 0
    j = 0
    print("------------------------------------------------------------------------")
    print("****** Merge Function ******")
    print(f"i\tleft\t\tj\tright\t\tresult")
    print("------------------------------------------------------------------------")
    while i < len(left) and j < len(right):
        print(f"{i}\t{left}\t{j}\t{right}\t{result}")
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    while i < len(left):
        result.append(left[i])
        i += 1
        print(f"{i}\t{left}\t{j}\t{right}\t{result}")
    while j < len(right):
        result.append(right[j])
        j += 1
        print(f"{i}\t{left}\t{j}\t{right}\t{result}")
    return result

left = [5,22,31,35]
right = [1,34,46]

result = merge(left,right)
````

### Ordenamiento por Mezcla Recursiva

Nótese que la complejidad del proceso iterativo recursivo es logarítmico, ya que en cada iteración, la lista se divide en dos. Además, en cada iteración debe hacerse una mezcla que, como se vio, tiene complejidad lineal. Es así como se concluye que la complejidad asintótica de `recursiveSort` es:

$$ O(\log(len(L))*len(L))$$



````python
def recursiveSort(L):
    if len(L) < 2:
        return L[:]
    else:
        mid = len(L)//2
        left = recursiveSort(L[:mid])
        right = recursiveSort(L[mid:])
        return merge(left,right)


L = [38,27,43,3,9,56,23,-54]

result = recursiveSort(L)
print(f"\n\nresult: {result}")
````

### Atributos públicos y privados 🔐

Por convención en Python:
- Atributos que **comienzan con `_` (guion bajo)** son considerados **protegidos** (no deberían usarse fuera de la clase).
- Atributos que **comienzan con `__` (doble guion bajo)** son **privados** (Python los oculta usando _name mangling_).

```python
self._curso = "Python"     # protegido (convenio)
self.__id = 12345          # privado (oculto)


````python
class Persona:
    def __init__(self, nombre, edad,curso = "Python"):
        self.nombre = nombre
        self.edad = edad
        self._curso = curso     # protegido (convenio)
        self.__id_curso = 12345          # privado (oculto)

    def print_id_curso(obj):
        print(obj.__id_curso)

    def get_id_curso(self):
        print("Llamando otro metodo en este:",end=" | ")
        self.print_id_curso()
        return self.__id_curso


pe = Persona("Cristian", 31)
print(pe._curso)
# pe.__id_curso
````

````python
# @title ### 🔷 CELDA CODE – Clase con públicos, privados y `__str__`

class Persona:
    def __init__(self, nombre, edad, id_secreto):
        self.nombre = nombre        # público
        self._ciudad = "Desconocida"  # protegido (convención)
        self.__id = id_secreto      # privado real (name mangling)
        self.edad = edad

    def __str__(self):
        return f"{self.nombre} ({self.edad} años)"

    def mostrar_id(self):
        return self.__id

# Crear objeto
p = Persona("Ana", 25, 9988)

# Acceder a públicos
print(p.nombre)
print(p.edad)

# Acceder a método __str__
print(p)

# Acceder a privado usando método de clase
print("ID secreto:", p.mostrar_id())

# Acceder directamente al atributo privado (NO recomendado)
# print(p.__id)            # Error
print(p._Persona__id)      # Funciona, pero rompe el encapsulamiento
````
