### 3.1 Condicionales

````python
a = input("Ingreseme un numero entero a:") # input solo retorna strings (str)
b = input("Ingreseme un numero entero b:") # input solo retorna strings (str)

a = int(a) # Convierte lo que tenga la variable "a" a entero (int|integer)
b = int(b) # Convierte lo que tenga la variable "b" a entero (int|integer)

# Saber quién es mayor a o b.

if a==b: #(5 == 6 --> False)
    print("a es igual a b")
    # print("--- Fin del bloque a==b ---") # Control
elif a<b: # elif = else if
    print("b es mayor a a")
    # print("--- Fin del bloque a<b ---")
else: # a>b
    print("a es mayor a b")
    # print("--- Fin del bloque ELSE ---")


print("--- Fin del if statement ---")

````

#### Bucle `while`:
El bucle `while` en Python permite ejecutar repetidamente un bloque de código mientras una condición específica sea verdadera. Este tipo de bucle es útil cuando no se conoce de antemano cuántas veces se debe repetir una acción.

Sintaxis del Bucle `while`:


```python
while condition: # condition es booleana, True o False
  # Bloque de código a ejecutar mientras la condición sea verdadera
#Fin del while




````python
contador = 1
N = 7
while contador < N:
    print("\tContador:", contador)
    contador = contador + 1 # contador += 1
print("<Fin While>")
print(contador*N)

# a=5
# b = 6
# print("a + b es:",a+b)
````

##### ```continue```

continue se hace uso en Loop, for loop o while loop, y sirve para ignorar parte de las instrucciones dispuestas en el mismo.

````python
# El usuario ingresa un valor N, y sólo se muestra los últimos 5 valores.

N = int(input("Ingrese N:")) #10
contador = 0
while contador < N: # 5 < 10
    contador = contador + 1 # 6
    if contador <= N-5: # 6 <= 5
        continue
    print(contador)
    print("------------")
print("<Fin del While>")
````

##### ```break```

Sentencia para "Romper" el ciclo, Finalizarlo.


````python
# Usuario ingresa numeros, decir si son pares o impares, si ingresa un número negativo debe finalizar

while True: # 0 == 0
    n = int(input("Ingrese valor:"))
    if n < 0:
        break
    if n%2 == 0:
        print("\tRes// el número es Par.")
    else:
        print("\tRes// el número es Impar.")

print("<Fin del While>")
````

### Funciones

Las **funciones** en Python son bloques de código reutilizables que realizan una tarea específica. Nos permiten organizar mejor el código, evitar repeticiones y hacerlo más legible y modular.

---

#### ¿Cómo se define una función?

Se utiliza la palabra clave `def`, seguida del nombre de la función, paréntesis con los parámetros (si los tiene) y dos puntos. El bloque de código dentro de la función se escribe indentado.

```python
def saludar():
    print("¡Hola, bienvenido a Python!")


````python
# Función sin argumento
def saludar():
    print("¡Hola, bienvenido a Python!")
saludar()
````

````python
# Uso repetitivo
for i in range(5):
    saludar()
````

````python
# Funciones con Parametro
def saludar(nombre):
    print(f"¡Hola!, {nombre}.")

saludar("Chicos")
````

````python
saludar()
````

````python
# Función con parámetro opcional
def saludar(nombre="Default Value"):
    print(f"¡Hola!, {nombre}.")


saludar("Chicos")
````

````python
saludar()
````

````python
# Función de más de un parámetro.
def suma_v1(a,b):
    return a+b

def suma_v2(a,b):
    sum = a + b
    return sum

sum_v1 = suma_v1(5,6)
sum_v2 = suma_v2(5,6)

print(f"Resultado suma_v1: {sum_v1}")
print(f"Resultado suma_v2: {sum_v2}")
````

````python
# Parámetros por posición
def show_info(nombre,edad):
    print(f"Nombre:{nombre}    |    Edad:{edad}")

print("\n********** Ejemplo asignación de parámetro por orden de parámetro **********\n")

show_info("Cristian",31)
"""
Se debe respetar el orden de los parámetros
esta es una forma como el traductor sabe qué
valor asigna a las parámetros de entrada de la función.
"""
show_info(31,"Cristian")


"""
Otra forma de decirle a la función qué valor asignar a
los parámetros de la misma, es definiendo a que parámetro de
la función se asigna el valor.
"""
print("\n********** Ejemplo asignación de parámetro de forma directa **********\n")
show_info(edad=31,nombre="Cristian")
show_info(nombre="Cristian",edad=31)

````
